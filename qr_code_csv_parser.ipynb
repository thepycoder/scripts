```python
import pandas as pd
import vobject
```

```python
contacts = pd.read_csv('qr_easy_scanner_history_3.csv')
contacts
```

```python
julie_rows = []
for row in contacts.itertuples():
    if not row.raw_value.startswith('BEGIN'):
        continue
    v = vobject.readOne(row.raw_value)
    v.prettyPrint()
    print(v.n.value.__dict__)
    try:
        title = v.title.value
    except:
        title = None
    julie_row = [v.n.value.given, v.n.value.family, v.email.value, v.org.value[0], title, row.notes]
    julie_rows.append(julie_row)
```

```python
julie_pd = pd.DataFrame(julie_rows, columns=['First Name', 'Family Name', 'email', 'Company Name', 'Title', 'Notes']).head(12)
julie_pd
```

```python
julie_pd.to_csv('julie_3.csv', index=None)
```
